### Receita = Quantidade * Valor_Unitario

```python
# Extrai o m√™s da fatura
df['Ano'] = df['Data Venda'].map(lambda date: date.year)
```

```python
df.head()
```

```python
# Calcular a receita
df["Receita"] = df["Quantidade"] * df["Pre√ßo Unit√°rio"]
```

```python
df.head()
```

```python
#Retornar maior faturamento
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
# Agrupa a receita por m√™s/ano
dados_receita = df.groupby(['Ano']).agg({'Receita': sum}).reset_index()
```

```python
dados_receita.head(10)
```

```python
dados_receita.tail(10)
```
